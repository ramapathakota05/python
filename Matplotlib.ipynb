## MatplotLib 

Matplotlib is a Python plotting library that works with NumPy, the numerical mathematics extension for Python. It offers an object-oriented API to integrate plots into applications using various GUI toolkits such as Tkinter, wxPython, Qt, or GTK+.

Key advantages of Matplotlib include:

Ease of use for creating simple plots
Ability to add custom labels and text
Extensive control over each element in a figure
High-quality output available in numerous formats
High level of customization

```python
#1st line imports the pyplot module from the 
#Matplotlib library

#2nd line is a magic command specific to Jupyter notebooks that configures 
#Matplotlib to display plots inline, 
#meaning the plots will appear directly below 
#the code cells that generate them

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
#scatter plotting example

x = np.arange(1,100,10)
y = x*10
plt.scatter(x,y, c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot example')
plt.savefig('scatter.png')
```

```python
## plt plot example
y=x/10
plt.plot(x,y,'r*-',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

y = x*2
#here we are saying 2 rows and 2 coumsn and 
#plot the image in position1
plt.subplot(2,2,1)
plt.plot(x,y,'r')

#here we are saying 2 rows and 2 coumsn and 
#plot the image in position2 and so on below
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

np.pi

```python
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
#np.pi is an attribute from the NumPy library that 
#represents the mathematical constant  (pi), 
#which is approximately 3.14159. 
#It is often used in mathematical calculations 
#involving circles, trigonometry, and other areas 
#where the value of pi is required.
np.pi
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
#here we are creating sine values using sine function in np
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
## Bar plot (seaborn produces better bar plots)

x1 = [2,8,10] 
y1 = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x1, y1) 
plt.bar(x2, y2) 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

## Histogram (you have a range of values, and it helps to understand what is the count or density that falls in the range or bins)

```python
a = np.array([22,30,10,50,43,56,73,80,54,15,20,40,65,79,31,27,70]) 
plt.hist(a, bins = 10) 
plt.title("histogram") 
plt.show()
```

## Box Plot 


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data);  #circles in below plot are the outliers
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210] #size here is based on cumulative total..total is around 800 and each value percentage  is calculated
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice (how much far the first slice to go)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
