Jointplot - helps you to perform univariant or bivariant analysis as it got 2 features... joinplot(x='dependent feature', y='independent feature',  data=df, kind = 'hex' or kind ='reg')..4 variables inside joinplot, dataframe of dependent and indepent features and then kde true or false

Pairplot.. A pair plot is a type of plot that shows the relationships between multiple features in a dataset by creating scatter plots for each pair of features. ex: sns.pairplot(df)
The hue parameter is commonly used in data visualization to add a categorical dimension to plots...so sns.pairplot(df, hue='select categorical feature')

distplot - helps you to check the distribution of the column features.
It combines a histogram and a Kernel Density Estimate (KDE) plot, providing a comprehensive view of the data distribution.sns.distplot(df['feature selection here'],kde=True,bins=10)
When you include KDE in the distplot, it overlays a smooth curve over the histogram. This curve represents the density of the data and helps in understanding the underlying distribution.
The histogram shows the count of data points in each bin.
The KDE plot (smooth curve) overlays the histogram, showing the estimated probability density function.
When you exclude KDE (df['feature selection here'],kde=False,bins=10), the distplot only shows the histogram, which gives a basic idea of the data distribution without the smooth density curve.
Only the histogram is shown, displaying the count of data points in each bin.
No smooth density curve is present.

Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. the following plots are covered:


countplot - A countplot in Seaborn is used to show the counts of observations in each categorical bin using bars. It's a simple and effective way to visualize the frequency distribution of categorical data. sns.countplot(data=iris, x='select column name here' and other axis by default will porivde the count) countplot created for selected column...this is a good example of univariant..sns.countplot(data=iris, x='select column name here', hue = also can be used to show the count of each combination of categories.

boxplot - is used to display the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (second quartile, Q2), third quartile (Q3), and maximum. It also highlights any outliers in the data. Box plots are particularly useful for comparing the distributions of different groups or categories.
sns.boxplot(x axis data, y axis data, data, hue parameter if required)

violinplot - A violin plot in Seaborn combines aspects of a box plot and a kernel density plot. It shows the distribution of the data across different categories, providing a richer understanding of the data distribution. The width of the violin indicates the density of the data at different values. sns.violinplot(x axis data, y axis data, data)

### Correlation with Heatmap

 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

```python
#seaborn has a property called as heatmap which help us
#to find out the inter relation between each and every feature.
#correlation can only be found when values are int or float..so including only int and float..
df_numeric = df.select_dtypes(include=[float, int])
df_numeric.corr()
#here you can see the corr bet bill and tip is 67% which means if your bill incresing tip also increasing
#this is good, but  if we want to visualise which are +vely correlated visually, which can
#be done using heatmap shown below cell..
```

```python
#in corr, values always range between -1 to +1, which is called as pearson coeff correlation
#heatmap helps us to visualise and all we have to do is the matrix of correlation in function

#this correlation is very much imp in EDA in feature selection..suppose there are 2 independent features
# and correlations bettwen those 2 are same, then we can use 1 feature instead of 2
sns.heatmap(df_numeric.corr())
```

## JoinPlot (helps you to do the univariant analysis, here we are taking 2 features, we can alsocalled as bivariant...make sure that you just have 2 features in joinplot)

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
### Univariate Analysis (can be called as bivariant as it got 2 features)




```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='hex')
#here we are taking which all features taking for the analysis..hex will display points in 
#hexagon shape instead of defauly scatter points

#below plot, you can see histograms plotted on both axis and major concentration of data is shown 
#where the bar is high..in between 10 and 20$..tips are between 2 to 4$..

#we can also see outlier where the greater bill 50$ and tip size also high
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='reg')
#lets play with kind parameter..instead of hex we are using reg..what it does is based on 
#histrogram, it will draw kde or pdf (prob dens function)...we can also see it draws a regression line
#or a best fiit line
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x='smoker',y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
